```python
!pip install kaleido

#Preprocessing the text file that contains information about training loss
import pandas as pd

file = open("/kaggle/input/d/ashishr9661/test-acc-shrec/testacc_log.txt") 
content = file.readlines() 

with open('output.txt', 'w') as f1:
    for i in range(1,len(content)-1,3):
        f1.write(content[i])

f1 = open('output.txt', 'r')
f2 = open('file2.txt', 'w')
checkWords = (":","[","]","\n","%")
repWords = (",","","","","\n")

for line in f1:
    for check, rep in zip(checkWords, repWords):
        line = line.replace(check, rep)
    f2.write(line)
f1.close()
f2 = open('file2.txt', 'r')
f2.readlines()

df = pd.read_csv('file2.txt', names=['Epoch1','Epoch','Test acc1', 'Test acc'])    

#pd.set_option('display.max_rows', None)
df.head()
```
